```python
import numpy as np, colour as cs

def isoluminant_sinebow(n=256, L=65, C=60):
    """Return n RGB triples forming an isoluminant sinebow."""
    θ = np.linspace(0, 2*np.pi, n, endpoint=False)
    h = np.degrees(θ)                          # LCh hue angle
    # Build LCh array, then convert to sRGB (assumes D65)
    lch = np.column_stack([np.full(n, L), np.full(n, C), h])
    rgb = cs.cam02ucs_to_srgb(cs.lch_to_cam02ucs(lch))
    return np.clip(rgb, 0, 1)

isoluminant_sinebow()
```

```python
from pathlib import Path
from collections import defaultdict
import numpy as np

root = Path('/Volumes/HiprDrive/hipr-screen/2023_10_24_lactobacilli_baselinecontrols')

# container for per-group sums
totals = defaultdict(int)

for seg_file in root.rglob("*host_seg.npy"):
    # first underscore-delimited token becomes the group label
    group = seg_file.stem.split('_')[0]

    arr = np.load(seg_file, mmap_mode="r")           # memory-map large masks
    n_labels = np.unique(arr).size                   # distinct integers (incl. background 0)
    totals[group] += n_labels                        # accumulate

# pretty-print sorted results
for grp, count in sorted(totals.items()):
    print(f"{grp}: {count}")

print("\nAs a dict:")
print(dict(totals))
```
